### Introduccion de los datos del "junior" en dos listas

```python
import pandas.io.excel
import numpy as np
Li=[]   # Lecturas
Lo=[]   # Ocupacion
for i in np.arange(0, 9, 1):
  Li.append(pandas.io.excel.read_excel("file://localhost/home/ubuntu/notebook/CopiaGasMensual.xls",
                                       sheetname=i, header=1, skip_rows=[0,1,2,3], parse_cols="K:L",
                                       skip_footer=45))
for i in np.arange(0, 9, 1):
  Lo.append(pandas.io.excel.read_excel("file://localhost/home/ubuntu/notebook/CopiaGasMensual.xls",
                                       sheetname=i, header=45, skip_rows=45, parse_cols="L:M",
                                       skip_footer=1))
```

```python
for i in np.arange(0, 9, 1):
    Lo[i]['CONSUMO']=0
    for key in Li[i].keys():
        j = 1
        while j < Lo[i]['OCUPACION'].count():
            Lo[i]['CONSUMO'][j] += Li[i][key][j+1]-Li[i][key][j]
            j += 1
```
